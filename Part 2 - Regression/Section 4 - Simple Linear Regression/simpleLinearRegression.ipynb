```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
dataset = pd.read_csv('Salary_Data.csv')
x = dataset.iloc[:, 0].values
y = dataset.iloc[:, 1].values

from sklearn.model_selection import train_test_split
xTrain, xTest, yTrain, yTest = train_test_split(x, y, test_size = 1 / 3, random_state = 0)

xTrain, yTrain, xTest, yTest = xTrain.reshape(-1, 1), yTrain.reshape(-1, 1), xTest.reshape(-1, 1), yTest.reshape(-1, 1)
```

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(xTrain, yTrain)
yPredicted = regressor.predict(xTest)
```

```python
plt.scatter(xTrain, yTrain, color = 'r', s = 10)
plt.plot(xTrain, regressor.predict(xTrain), color = 'b', linewidth = 0.5)
plt.title('Salary vs Exprience (Training set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```

```python
plt.scatter(xTest, yTest, color = 'r', s = 10)
plt.plot(xTrain, regressor.predict(xTrain), color = 'b', linewidth = 0.5)
plt.title('Salary vs Exprience (Test set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```

```python
regressor.score(xTest, yTest)
```
