```python
# Polynomial Regression
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# Importing the dataset
dataset = pd.read_csv('Position_Salaries.csv')
x = dataset.iloc[:, 1:2].values
y = dataset.iloc[:, 2].values
```

```python
plt.scatter(x, y, s = 30, c = 'r')
plt.show()
```

```python
# Fitting Linear Regression to the dataset
from sklearn.linear_model import LinearRegression
linearRegressor = LinearRegression()
linearRegressor.fit(x, y)
```

```python
# Fitting Polynomial Regression to the dataset
from sklearn.preprocessing import PolynomialFeatures
polynomialFeature = PolynomialFeatures(5)
xPoly = polynomialFeature.fit_transform(x)
```

```python
polynomialRegressor = LinearRegression()
polynomialRegressor.fit(xPoly, y)
```

```python
# Visualizing the Linear Regression
plt.scatter(x, y, s = 30, c = 'r')
plt.plot(x, linearRegressor.predict(x), c = 'b')
plt.title('Truth or Bluff (Linear Regression)')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show()
print('The accuracy of linear regression is {}'.format(linearRegressor.score(x, y) * 100))
```

```python
# Visualizing the Polynomial Regression
plt.scatter(x, y, s = 30, c = 'r')
plt.plot(x, polynomialRegressor.predict(xPoly), c = 'b')
plt.title('Truth or Bluff (Linear Regression)')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show()
print('The accuracy of polynomial regression is {}'.format(polynomialRegressor.score(xPoly, y) * 100))
```
