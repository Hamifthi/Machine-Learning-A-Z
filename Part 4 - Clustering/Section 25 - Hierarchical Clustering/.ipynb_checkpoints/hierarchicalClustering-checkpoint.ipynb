```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy.cluster.hierarchy as sch
```

```python
dataset = pd.read_csv('Mall_Customers.csv')
x = dataset.iloc[:, [3, 4]].values
```

```python
# using Dendrogram
dendrogram = sch.dendrogram(sch.linkage(x, method = 'ward'))
plt.title('Dendorgram')
plt.xlabel('Customers')
plt.ylabel('Euclidean distances')
plt.show()
```

```python
# Fitting hierarchical clustering to the dataset
from sklearn.cluster import AgglomerativeClustering
hc = AgglomerativeClustering(n_clusters = 5, affinity = 'euclidean', linkage = 'ward')
yHC = hc.fit_predict(x)
```

```python
# Visiualising the clusters
colorsList = ['red', 'blue', 'green', 'cyan', 'magenta']
labelList = ['Careful', 'Standard', 'Target', 'Careless', 'Sensible']
for number in range(5):
    plt.scatter(x[yHC == number, 0], x[yHC == number, 1], s = 50,
    c = colorsList[number], label = 'cluster {}'.format(labelList[number]))
plt.title('Cluster of clients')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending score (1-100)')
plt.xlim(xmin = 0, xmax = 230)
plt.legend()
plt.show()
```
