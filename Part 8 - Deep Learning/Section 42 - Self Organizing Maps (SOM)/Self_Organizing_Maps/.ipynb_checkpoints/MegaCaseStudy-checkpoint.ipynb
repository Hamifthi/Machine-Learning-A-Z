```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv('dataset')
x = dataset.iloc[:, 2:-1].values
y = dataset.iloc[:, -1].values
```

```python
from sklearn.preprocessing import MinMaxScaler
sc = MinMaxScaler()
x = sc.fit_transform(x)
```

```python
from sklearn.model_selection import train_test_split
xTrain, xTest, yTrain, yTest = train_test_split(x, y, test_size = 0.2, random_state = 0)
```

```python
import keras
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import Dropout
```

```python
model = Sequential()
```

```python
model.add(Dense(activation = 'relu', input_dim = 14, units = 6, kernel_initializer = 'uniform'))
model.add(Dropout(rate = 0.2))
```

```python
model.add(Dense(activation = 'relu', units = 6, kernel_initializer = 'uniform'))
model.add(Dropout(rate = 0.2))
```

```python
model.add(Dense(activation = "sigmoid", units = 1, kernel_initializer = 'uniform'))
```

```python
model.compile(loss = 'binary_crossentropy', optimizer = 'adam', metrics = ['accuracy'])
```

```python
model.fit(xTrain, yTrain, batch_size = 5, epochs = 200)
```

```python
yPredicted = model.predict(xTest)
yPredicted = yPredicted > 0.5
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(yTest, yPredicted)
print(cm)
```
