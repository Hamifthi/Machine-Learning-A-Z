```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv('Credit_Card_Applications.csv')
x = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
```

```python
from sklearn.preprocessing import MinMaxScaler
sc = MinMaxScaler(feature_range = (0, 1)) 
x = sc.fit_transform(x)
```

```python
from minisom import MiniSom
som = MiniSom(x = 10, y = 10, input_len = 15, random_seed = 10)
som.random_weights_init(x)
som.train_random(x, num_iteration = 100)
```

```python
from pylab import bone, pcolor, colorbar, plot, show
bone()
pcolor(som.distance_map().T)
colorbar()
markers = ['o', 's']
colors = ['r', 'g']
for i, j in enumerate(x):
    w = som.winner(j)
    plot(w[0] + 0.5,
         w[1] + 0.5,
         markers[y[i]],
         markeredgecolor = colors[y[i]],
         markerfacecolor = "None",
         markersize = 10,
         markeredgewidth = 2)
show()
```

```python
mappings = som.win_map(x)
frauds = np.concatenate((mappings[(2, 1)], mappings[(2, 2)], mappings[(1, 5)], mappings[(6, 1)]), axis = 0)
frauds = sc.inverse_transform(frauds)
frauds.shape
```

```python
listOfY = []
for i in frauds[:, 0]:
    for j in range(len(dataset)):
        if i == dataset.iloc[j, 0]:
            listOfY.append(dataset.iloc[j, 15])
```

```python
listOfY = np.array(listOfY).reshape(71, 1)
frauds = np.hstack((frauds, listOfY))
```

```python
dataset = pd.DataFrame(frauds).to_csv('dataset')
```
